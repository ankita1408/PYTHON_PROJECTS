```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\ankit\Downloads\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
##df['Last_Name']=df['Last_Name'].str.strip('...')
##df['Last_Name']=df['Last_Name'].str.strip('/')
##df['Last_Name']=df['Last_Name'].str.strip('_')
##df['Last_Name']
```

```python
df
```

```python
df['Last_Name']=df['Last_Name'].str.strip('123._/')
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df['Phone_Number']
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('|','')
df['Phone_Number']
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('/','')
df['Phone_Number']
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('-','')
df['Phone_Number']
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']
```

```python
df['Phone_Number']=df['Phone_Number'].apply(
    lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df['Phone_Number']
```

```python
df
```

```python
df[['street_address','state','zip_code']]=df['Address'].str.split(',',n= 2, expand=True)
df[['street_address','state','zip_code']]
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Paying Customer']=df['Paying Customer'].str.replace('N/a','')
df['Paying Customer']
```

```python
df=df.replace('N/a','')
df=df.fillna('')
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':  # Fixed syntax
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':  # Fixed syntax
        df.drop(x, inplace=True)
df
```
